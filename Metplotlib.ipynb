## Matplotlib
Matplotlib is an open-source Python library for creating a wide range of visualizations, from simple graphs to complex plots, making it valuable in data analysis and presentation.

```python
# To importmatplotlib
import matplotlib.pyplot as plt # pyplot is another library inside matplotlib
%matplotlib inline 
# this will show the diagrams just next to the code else everytime user would have to write plt.show()
```

```python
import numpy as np
x = np.arange(0,10)
y = np.arange(11,21)
```

### Scatter plot ---
A scatter plot is a graphical representation used to display the relationship between two variables, showing individual data points as dots on a two-dimensional plane.

```python
# To use x and y and create scatter plot --- The array size should be same for both cordinates x and y
plt.scatter(x,y)
```

```python
# To change color of scatter plot and add details like labels, title
plt.scatter(x,y,c="g")
plt.xlabel('X Label')
plt.ylabel('Y Label') # labels should be used in single code line like this
plt.title('Graph in 2D')
# To save the fig size which is (0.5,1.0, 'Graph in 2D') for current plot use ---
plt.savefig('test.png')
```

```python
# To create line plot
y = x*x
plt.plot(x,y)
```

```python
# To change color and the shape of line plot and add other things 
plt.plot(x,y,'g*',linestyle='dashed',linewidth=1.5,markersize=12.5)
#Marker size is result points 
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('2D Diagram')
plt.savefig('test1.png')
```

### Subplots
A subplot in Matplotlib refers to a grid of smaller plots within a single larger plot, allowing multiple visualizations to be displayed together.

```python

plt.subplot(2,2,1)
plt.plot(x,y,'g--')
plt.title('2,2,1 Diagram')
plt.subplot(2,2,2)
plt.plot(y,x,'r--')
plt.title('2,2,2 Diagram')
plt.subplot(2,2,3)
plt.plot(y,x,'y--')
plt.xlabel('2,2,3 Diagram')
plt.subplot(2,2,4)
plt.plot(x,y,'b--')
plt.xlabel('2,2,4 Diagram')
```

### Bar plots
Bar plots in data visualization are graphical representations that use rectangular bars to show the relationship between different categories or groups, making it easy to compare quantities or values.

```python
x1=[2,3,4,5]
y1=[11,16,19,20]

x2=[3,9,11,14]
y2=[6,15,7,9]
plt.bar(x1,y1)
plt.bar(x2,y2,color='g')
plt.title('Bar plot')
plt.xlabel('X array')
plt.ylabel('Y array')
```

### Histogram
A histogram in Matplotlib is a graphical representation that uses bars to show the distribution of a dataset, with each bar representing a range of values and the height of the bar indicating the frequency or count of values within that range.

```python
a = [22,4,55,33,2,12,66,83,2,1,54,7,32,77,15]
plt.hist(a)
```

#### This shows the frequency of the elements like in above histogram chart number of elements below till 10 are 5 hence it is creating a bar at 10 on x axis and 5 on y axis

### Box Plot
A box plot in Matplotlib is a visualization that displays the distribution of a dataset using a rectangular "box" that represents the interquartile range (IQR) of the data, along with "whiskers" that extend to show the data range within a certain range of values. It helps to understand the central tendency and variability of the data, as well as identify potential outliers.

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data)
```

### Pie Chart
A pie chart is a circular graphical representation that divides data into slices or sectors, with each slice representing a portion of the whole. The size of each slice is proportional to the quantity it represents, making it easy to compare relative proportions or percentages of different categories within the data.

```python
data=[23,17,29,40,50,41]
cars = ['AUDI','BMW','FORD','JAGUAR','TESLA','MERCEDES']
explode = [0.1,0.0,0.2,0.3,0.0,0.0]
colors = ['orange','cyan','brown','grey','indigo','beige']
wp={'linewidth':1,'edgecolor':"green"}
plt.pie(data,labels=cars,colors=colors,explode=explode,wedgeprops=wp)
```
